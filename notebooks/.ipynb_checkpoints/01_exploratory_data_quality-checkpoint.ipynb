# 2.3 Conteo de nulos

```python
def missing_report(df):
    return (
        df.isna()
        .sum()
        .to_frame("missing_count")
        .assign(missing_pct=lambda x: x["missing_count"] / len(df))
        .sort_values("missing_pct", ascending=False)
    )

missing_report(customers)
missing_report(accounts)
missing_report(transactions)
```

# 2.6 Validación Relacional

```python
# Accounts con customer inexistente
invalid_customer_accounts = accounts[
    ~accounts["customer_id"].isin(customers["customer_id"])
]

len(invalid_customer_accounts)
```

```python
# Transactions con account inexistente
invalid_account_transactions = transactions[
    ~transactions["account_id"].isin(accounts["account_id"])
]

len(invalid_account_transactions)
```

# 2.7 Validación Financiera

```python
# Amount como string
transactions["amount"].apply(type).value_counts()
```

```python
# Inconsistencia signo vs tipo
transactions["amount_numeric"] = pd.to_numeric(
    transactions["amount"], errors="coerce"
)

inconsistent = transactions[
    ((transactions["transaction_type"] == "debit") &
     (transactions["amount_numeric"] > 0)) |
    ((transactions["transaction_type"] == "credit") &
     (transactions["amount_numeric"] < 0))
]

len(inconsistent)
```

```python
# Outliers extremos
transactions["amount_numeric"].describe()
```

```python
plt.figure()
transactions["amount_numeric"].hist(bins=100)
plt.title("Distribution of Transaction Amounts")
plt.show()
```

```python
# Fechas futuras
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"], errors="coerce"
)

future_transactions = transactions[
    transactions["transaction_date"] > pd.Timestamp.today()
]

len(future_transactions)
```

# 2.8 Validación de Emails

```python
email_pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

invalid_emails = customers[
    ~customers["email"].astype(str).str.match(email_pattern)
]

len(invalid_emails)
```

# 2.9 Score Básico de Calidad

```python
quality_metrics = {
    "duplicate_accounts": accounts["account_id"].duplicated().sum(),
    "duplicate_transactions": transactions["transaction_id"].duplicated().sum(),
    "invalid_customer_links": len(invalid_customer_accounts),
    "invalid_account_links": len(invalid_account_transactions),
    "future_transactions": len(future_transactions),
    "invalid_emails": len(invalid_emails),
}

pd.Series(quality_metrics)
```
